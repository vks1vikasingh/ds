```python
import chardet
import pandas as pd;
import numpy as np
```

```python
companies = pd.read_csv('companies.txt',sep='\t',encoding="ISO-8859-1")
companies.head()
```

```python
# We noticed there are some encoding issues that we removed as below
companies['permalink'] = companies[['permalink']].applymap(lambda x: x.encode('utf-8').decode('ascii', 'ignore'))
```

```python
companies.head(3)
```

```python
rounds2 = pd.read_csv('rounds2.csv',encoding="ISO-8859-1")
rounds2.head(3)
```

```python
# We noticed there are some encoding issues that we removed as below
rounds2['company_permalink'] = rounds2[['company_permalink']].applymap(lambda x: x.encode('utf-8').decode('ascii', 'ignore'))
```

```python
rounds2.head(3)
```

```python
# How many unique companies are present in rounds2?
rounds2['company_permalink'].str.upper().nunique()
```

```python
# How many unique companies are present in the companies file?
companies['permalink'].str.upper().nunique()
```

```python
#In the companies data frame, which column can be used as the  unique key for each company? Write the name of the column.
# we checked to unique count and normal count. 
print('Normal count',companies['permalink'].str.upper().count())
print('Unique count',companies['permalink'].str.upper().nunique())
```

```python
#Are there any companies in the rounds2 file which are not  present in companies ? Answer Y/N.
# After resolving encoding issue we found count is same in rounds2 company_permalink and companies permalink.
```

```python
#Merge the two data frames so that all  variables (columns)  in the companies frame are added to the rounds2 data frame.
master_frame = companies.merge(rounds2, left_on=companies['permalink'].str.upper(), right_on=rounds2['company_permalink'].str.upper())
```

```python
print(master_frame['permalink'].str.upper().nunique())
print(master_frame['company_permalink'].str.upper().nunique())
```

```python
master_frame.head(3)
```

```python
#total count in master_frame
master_frame.count()
```

```python
# To display mean in non exponential format we used below
pd.set_option('display.float_format', lambda x: '%.2f' % x)
```

```python
#Table 2.1 (Average funding amount of all funding_round_type)
print(master_frame.groupby('funding_round_type')['raised_amount_usd'].mean())
```

```python
#Table-2.1  we can see venture investment ranges betwwen 5 million to 15 million
master_frame.dropna(subset=['raised_amount_usd'],inplace = True)
print(master_frame.groupby('funding_round_type')['raised_amount_usd'].mean())
```

```python
# For country analysis we will use venture_master_frame data frame
venture_master_frame = master_frame[master_frame['funding_round_type']=='venture']
venture_master_frame.head(3)
```

```python
# make a data frame named top9 with the top nine countries (based on the total investment amount each country has received)
top9 = venture_master_frame.groupby('country_code')['raised_amount_usd'].sum().sort_values(ascending=False).head(9)
top9
```

```python
#Checkpoint 4: Sector Analysis 1 
```

```python
# creating the priamry sector by tokenzing the 
venture_master_frame['primary_sector']= venture_master_frame["category_list"].str.split("|", n = 1, expand = True)[0]
```

```python
venture_master_frame.head(3)
```

```python
mapping = pd.read_csv('mapping.csv',encoding="ISO-8859-1")
mapping.head()
```

```python
mapping.count()
```

```python
# repacing the null category_list with BLANK for proper JOIN
mapping[mapping['category_list'].isnull()].count()
mapping['category_list'].fillna(value='BLANK',inplace=True)
```

```python
# transpose the column into row to join with master data frame
mapping_melt=pd.melt(mapping, id_vars =['category_list'], value_vars =['Automotive & Sports','Blanks','Cleantech / Semiconductors','Entertainment','Health','Manufacturing','News, Search and Messaging','Others','Social, Finance, Analytics, Advertising'])
mapping_melt_1 = mapping_melt[mapping_melt['value']==1]
```

```python
venture_master_frame['category_list'].fillna(value='BLANK',inplace=True)
venture_master_frame = venture_master_frame.merge(mapping_melt_1, left_on=venture_master_frame['primary_sector'].str.upper(), right_on=mapping_melt_1['category_list'].str.upper())
```

```python
venture_master_frame.head(3)
```

```python
#  merged data frame with each primary sector mapped to its main sector
venture_master_frame= venture_master_frame.rename(index=str, columns={"variable": "main_sector","category_list_x":"category_list"})
```

```python
# the primary sector and main sector present in a separate column).
venture_master_frame.drop(['category_list_y','value'],axis=1,inplace=True)
```

```python
venture_master_frame.head(3)
```

```python
# Top three English speaking countries and the most suitable funding type for Spark Funds is 'venture'.
country1 = venture_master_frame[(venture_master_frame.country_code == 'USA') & (venture_master_frame.funding_round_type == 'venture') ]
country2 = venture_master_frame[(venture_master_frame.country_code == 'GBR') & (venture_master_frame.funding_round_type == 'venture') ]
country3 = venture_master_frame[(venture_master_frame.country_code == 'IND') & (venture_master_frame.funding_round_type == 'venture') ]
```

```python
D1=country1[(5000000 <=  country1['raised_amount_usd']) &  (country1['raised_amount_usd']  <=  15000000)]
D1.head(3)
```

```python
D2=country2[(5000000 <=  country2['raised_amount_usd']) &  (country2['raised_amount_usd']  <=  15000000)]
D2.head(3)
```

```python
D3=country3[(5000000 <=  country3['raised_amount_usd']) &  (country3['raised_amount_usd']  <=  15000000)]
D3.head(3)
```

```python
D1['Total_Investment'] = D1.groupby('main_sector')['raised_amount_usd'].transform('sum')
D1['Total_count'] = D1.groupby('main_sector')['raised_amount_usd'].transform('count')
D1.tail(3)
```

```python
D2['Total_Investment'] = D2.groupby('main_sector')['raised_amount_usd'].transform('sum')
D2['Total_count'] = D2.groupby('main_sector')['raised_amount_usd'].transform('count')
D2.tail(3)
```

```python
D3['Total_Investment'] = D3.groupby('main_sector')['raised_amount_usd'].transform('sum')
D3['Total_count'] = D3.groupby('main_sector')['raised_amount_usd'].transform('count')
D3.tail(3)
```
